```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('C:/users/skippy/documents/ncaa2008.csv',usecols=['Winner','Loser'],encoding='iso8859_15')
print(df)
```

```python
G = nx.from_pandas_edgelist(df,source='Loser',target='Winner',create_using=nx.DiGraph())
```

```python
nx.draw_spring(G,node_size=10)
plt.show()
```

```python
sorted(nx.eigenvector_centrality(G).items(), key=lambda kv: kv[1], reverse=True)
```

```python
sorted(nx.in_degree_centrality(G).items(), key=lambda kv: kv[1], reverse=True)
```

```python
sorted(nx.closeness_centrality(G).items(), key=lambda kv: kv[1], reverse=True)
```

```python
sorted(nx.pagerank(G).items(), key=lambda kv: kv[1], reverse=True)
```

```python
from networkx.algorithms import community
C = community.kernighan_lin_bisection(G)
```
